```python
# Convertir les colonnes de date en datetime
df_entre['date'] = pd.to_datetime(df_entre['date'])
df_movie['date_sortie'] = pd.to_datetime(df_movie['date_sortie'])

# Puis compter le nombre de films en 2005
df_entre_2005 = df_entre[df_entre['date'].dt.year == 2014]
print(f"Nombre de films en 2005 dans df_entre : {len(df_entre_2005)}")

df_movie_2005 = df_movie[df_movie['date_sortie'].dt.year == 2014]
print(f"Nombre de films en 2005 dans df_movie : {len(df_movie_2005)}")

```

```python
# enleve les : et les espaces
df_entre['title'] = df_entre['title'].str.replace(':', '')
df_movie['titre_fr'] = df_movie['titre_fr'].str.replace(':', '')
# enleve les espaces en trop de tous
df_entre['title'] = df_entre['title'].str.replace(' ', '')
df_movie['titre_fr'] = df_movie['titre_fr'].str.replace(' ', '')
#met tous en minuscule
df_entre['title'] = df_entre['title'].str.lower()
df_movie['titre_fr'] = df_movie['titre_fr'].str.lower()
# enleve les accents titre
df_entre['title'] = df_entre['title'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df_movie['titre_fr'] = df_movie['titre_fr'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
# enleve l'accent '
df_entre['title'] = df_entre['title'].str.replace("'", '')
df_movie['titre_fr'] = df_movie['titre_fr'].str.replace("'", '')
# enleve virgurle
df_entre['title'] = df_entre['title'].str.replace(",", '')
df_movie['titre_fr'] = df_movie['titre_fr'].str.replace(",", '')
#enleve tiret
df_entre['title'] = df_entre['title'].str.replace("-", '')
df_movie['titre_fr'] = df_movie['titre_fr'].str.replace("-", '')
```

```python
#enleve les espaces tous les espaces
df_entre['title'] = df_entre['title'].str.replace(" ", '')
df_movie['titre_fr'] = df_movie['titre_fr'].str.replace(" ", '')
```

```python
# Renommer les colonnes de df2 pour correspondre à celles de df1
df2 = df_movie.rename(columns={'titre_fr': 'title', 'date_sortie': 'date'})
```

```python
# Fusionner les deux dataframes sur le titre et la date
df = pd.merge(df_entre, df2, on=['title', 'date'])

# Calculer le nombre de films qui n'ont pas pu être fusionnés
num_unmerged = len(df_entre) + len(df2) - len(df)

print(f"Nombre de films qui n'ont pas pu être fusionnés : {num_unmerged}")
```

```python
query = """
SELECT films.titre_non_modifie
FROM films 
JOIN films_acteurs ON films.film_id = films_acteurs.film_id 
JOIN acteurs ON films_acteurs.acteur_id = acteurs.acteur_id 
WHERE acteurs.nom = 'Dany Boon'
"""
cursor = connection.cursor()
cursor.execute(query)
results = cursor.fetchall()

# Convertir les résultats en DataFrame
df_films_dwayne = pd.DataFrame(results, columns=['titre_non_modifie'])
```

```python
query = """
SELECT films.titre_non_modifie
FROM films 
JOIN films_genres ON films.film_id = films_genres.film_id 
JOIN genres ON films_genres.genre_id = genres.genre_id 
WHERE genres.nom = 'Horreur'
"""
cursor = connection.cursor()
cursor.execute(query)
results = cursor.fetchall()

# Convertir les résultats en DataFrame
df_films_fantastique = pd.DataFrame(results, columns=['titre_non_modifie'])
df_films_fantastique
```

```python
query = """
SELECT films.titre_non_modifie
FROM films 
JOIN films_compagnies ON films.film_id = films_compagnies.film_id 
JOIN compagnies ON films_compagnies.compagnie_id = compagnies.compagnie_id 
WHERE compagnies.nom = 'Marvel Studios'
"""
cursor = connection.cursor()
cursor.execute(query)
results = cursor.fetchall()

# Convertir les résultats en DataFrame
df_films_dwayne = pd.DataFrame(results, columns=['titre_non_modifie'])
```
